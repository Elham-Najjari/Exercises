```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

pi = np.pi
x = np.linspace(-4*pi , 4*pi, 1000)
plt.plot(x, np.sin(x)/x)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

x = np.arange(0, 10.001, 0.25)
y = np.sin(x)
z = np.cos(x)

plt.plot(x, y, 'o-', color = 'red', label = 'f=sin')
plt.plot(x, z, 's-', color = 'blue', label = 'f=cos')

plt.legend()
#plt.show()

plt.xlabel ('x')
plt.ylabel('f(x)')
plt.grid(True)
```

```python
from sklearn.datasets import load_boston
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

boston = load_boston()
x = boston['data']
F = boston['feature_names']
#print(F)

x = x[: , [5,12]]

plt.figure(figsize = (6,6))
plt.plot(x[:,0], x[:,1], 'o' , color = 'pink' , ms=3)
plt.plot(x[:,0].mean(), x[:,1].mean(), 'o', color = 'cyan', ms = 10)
```
